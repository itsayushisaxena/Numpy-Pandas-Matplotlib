##### Preprocessing

```python
df.hp = df.hp.replace("?", np.nan)              
df.hp.mean()
```

```python
df.hp = df.hp.replace(np.nan, df.hp.mean())
df.dtypes                           #now,check the datatype of hp column
```

```python
# I want a car that has above average hp and mpg is above average
df[(df.mpg > df.mpg.mean()) & (df.hp > df.hp.mean())]
```

```python
# How does the average mpg value of all the cars change over the years
df[df.year == 82].head()     #cars manufactured in year 82
```

```python
# We'll use groupby here  *very useful*
res = df.groupby('year').mean().mpg
```

```python
import matplotlib.pyplot as plt
plt.plot(res)
plt.show()
```
